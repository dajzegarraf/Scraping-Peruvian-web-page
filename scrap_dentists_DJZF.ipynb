```python
import os
import time
from selenium import webdriver
import pandas as pd
import numpy as np
# When downloading chromedriver, it is important to point out its location on the computer to activate it
chromedriver = "C:/Users/User-pc/chromedriver_win32/chromedriver"

os.environ["webdriver.chrome.driver"] = chromedriver
```

```python
# chromedriver activation
b = webdriver.Chrome(chromedriver)
    
odonto_list=list()
# creating iteration according to number of the Dentist College
# range( from number (X), to number (Y-1)  )
for m in range(1000,47139):           # number of registered dentists on 19/05/20
            # selecting the web page from which the data will be extracted
            b.get("http://190.117.127.188/consultas_web/consulta_colegiado.asp")
            
            time.sleep(1)
            # enter the member number
            intro_num=b.find_element_by_xpath("/html/body/form/div/input[1]")
            intro_num.send_keys(m)
            
            # search the number
            buscar_num=b.find_element_by_xpath("/html/body/form/div/input[2]")
            buscar_num.click()
            
            # Medical College code
            codigo_odont=b.find_element_by_xpath("//table/tbody/tr[2]/td[1]").text
            
            # last names and names
            nom_completo=b.find_element_by_xpath("//table/tbody/tr[2]/td[3]").text
            
            ### create a conditional to go to the link and extract info on the next page if consejo_reg=LIMA.
            
            # consejo regional
            consejo_reg=b.find_element_by_xpath("//table/tbody/tr[2]/td[4]").text
            
            if consejo_reg=="LIMA":
                # state that takes you to another link
                # state = b.find_element_by_xpath("//table/tbody/tr[2]/td[5]/a").text
                estado_link=b.find_element_by_xpath("//table/tbody/tr[2]/td[5]/a").click()
                time.sleep(1)
                try: 
                    estado=b.find_element_by_xpath("html/body/section[2]/section[2]/div/div/div/article/table/tbody/tr/td[4]").text
                except:
                    estado=np.NaN  
            else:
                # activity status
                estado=b.find_element_by_xpath("//table/tbody/tr[2]/td[5]/div").text  
            
            # creating a dictionary   
            colegio_odonto=pd.Series({"Codigo COP":codigo_odont,"Nombre completo":nom_completo,
                         "Consejo Regional":consejo_reg,"Estado":estado})
            odonto_list.append(colegio_odonto)
            
# converting to a dataframe
df=pd.DataFrame(odonto_list)
b.close()                 
```
